Preprocessing Matters

Preprocessing is critical for LDA because the model is based on word distributions.
If we do not remove stop words:
- Common words like "the", "is" will dominate topics
- Topics will become meaningless

If we do not lemmatize:
- Words like "run", "running", "ran" will be treated as different words

LDA will produce noisy and unclear topics.

3. LDA vs Clustering

K-Means:
- Hard assignment → one document = one cluster

LDA:
- Soft assignment → one document = mixture of topics

A document can be- 70% technology, 30% medicine

```python
documents = [
    "God is love and faith is important in religion",
    "Atheism is the absence of belief in gods",
    "3D graphics and computer rendering are used in design",
    "Medical doctors study diseases and human health",
    "Computer graphics help visualize complex data",
    "Religion involves worship and spiritual beliefs",
    "Medicine helps treat illness and improve health",
    "Atheists do not believe in religion or gods",
    "Rendering images requires powerful graphics hardware",
    "Doctors diagnose and treat patients in hospitals"
]

#  Note
#The original dataset (20 Newsgroups) could not be loaded due to slow internet connection or long download time.
#Therefore, a small custom dataset was created manually to demonstrate how LDA topic modeling works.
#This does not affect the goal of the exercise, since the focus is on understanding and interpreting topics rather than working with large-scale data.
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer(
    max_df=0.95,
    min_df=2,
    stop_words='english'
)

doc_term_matrix = vectorizer.fit_transform(documents)
```

```python
from sklearn.decomposition import LatentDirichletAllocation

lda = LatentDirichletAllocation(
    n_components=4,
    random_state=42,
    n_jobs=-1
)

lda.fit(doc_term_matrix)
```

```python
feature_names = vectorizer.get_feature_names_out()

for topic_idx, topic in enumerate(lda.components_):
    print(f"\nTopic {topic_idx}:")
    
    top_words = topic.argsort()[-10:]
    
    print([feature_names[i] for i in top_words])
```

Topics should roughly correspond to:

1. Religion (God, church, faith)
2. Atheism (belief, argument, religion)
3. Graphics (image, file, computer)
4. Medicine (health, doctor, disease)

LDA may mix topics slightly because it is unsupervised.

```python
new_doc = [
    "A new medical imaging technique uses advanced computer graphics to visualize the human brain"
]

new_doc_vector = vectorizer.transform(new_doc)
```

```python
topic_distribution = lda.transform(new_doc_vector)

print(topic_distribution)
```

The output is a probability distribution over topics.
Example- [0.1, 0.05, 0.6, 0.25]

- Highest topic = Graphics (0.6)
- Second = Medicine (0.25)

The sentence contains both:
- "medical", "brain" → medicine
- "graphics", "visualize" → graphics

This shows LDA can detect mixed topics.
