Exercise1— Conceptual Questions

To generate clothing images conditioned on type and color, the condition vector y should encode both attributes jointly.

A practical design is:

Use a one-hot encoding for clothing type:
shirt = [1,0,0]
trousers = [0,1,0]
shoes = [0,0,1]
Use a one-hot encoding for color:
red = [1,0,0]
green = [0,1,0]
blue = [0,0,1]

Then concatenate them:
y = [clothing_type_one_hot || color_one_hot]

For example:
"blue shirt" → [1,0,0, 0,0,1]
This condition vector is fed into:

Generator (to control output)
Discriminator (to verify image matches condition)
CycleGAN Failure Modes
a) Possible failure mode
Without cycle consistency, the Generator could ignore the input image entirely and map all summer images to a single “typical winter scene” (e.g., snowy forest), regardless of content
This is called mode collapse or content ignoring, where the output is plausible but not related to the input

b) Role of cycle consistency loss
Cycle consistency forces:
summer → winter → summer ≈ original summer image


This prevents the model from ignoring input structure because:

If it discards content, it cannot reconstruct the original image
Therefore, it must preserve spatial and structural information
This enforces a one-to-one mapping between domains while still allowing style transformation
StyleGAN and Bias
If the training dataset contains demographic bias (e.g., overrepresentation of certain ethnicities or age groups), StyleGAN will replicate that bias in generated images.

This means:

The model will generate more faces resembling dominant groups in the dataset
Underrepresented groups will appear less frequently or with lower quality

Ethical concern:

This is significant because:
It reinforces societal biases in synthetic data
It can misrepresent diversity in applications like media, games, or datasets
It can unintentionally exclude or distort minority representation

Generative models amplify whatever is present in the training data, making dataset bias a direct output bias.

Exercise 2 — Designing a GAN Application

The best starting point for a sketch-to-art system is a Conditional GAN (cGAN).

Why
Input sketch acts as a conditioning signal
Generator learns mapping: sketch → realistic image
Discriminator checks if output is realistic AND matches sketch

CycleGAN is less suitable because it is designed for unpaired domain translation, not strict input-output control. StyleGAN is not ideal because it focuses on latent space generation, not direct input conditioning.

Dataset Requirements

A paired dataset is required:
Input: hand-drawn or synthetic sketches
Target: corresponding high-quality concept art

Each pair must align structurally:
(sketch, final artwork)

Without paired data, the model cannot learn exact mapping from sketch to final image.


1. Sketch variability
Players draw sketches with different skill levels:
messy lines
incomplete shapes
abstract forms
This makes input distribution highly noisy

2. Ambiguity of meaning
A sketch may be unclear:
a circle could be a head, a shield, or a planet
The model must infer intent, not just geometry

3. Style consistency
Maintaining the game’s art style while respecting sketch structure is difficult, especially across diverse inputs
