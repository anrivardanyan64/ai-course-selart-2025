Exercise 1- Conceptual Questions

1. Bounding Box Formats

(x_min=10, y_min=20, x_max=60, y_max=80):

x_center = (10 + 60) / 2 = 35  
y_center = (20 + 80) / 2 = 50  
width = 60 - 10 = 50  
height = 80 - 20 = 60  

Final format: (35, 50, 50, 60)


2. IoU for Different Shapes

The IoU will be lower for the trapezoid-shaped car
This is because a rectangular bounding box cannot tightly fit irregular shapes, so it includes more background area

This is a limitation of standard bounding boxes — they cannot accurately represent complex shapes


*3. The RPN's Job

The RPN only needs to suggest possible object locations (regions). 
It is okay if they are not precise because the second stage will refine them
The second stage classifies the object and adjusts the bounding box to be more accurate

Exercise 2- IoU Calculation

Ground Truth box:
(20, 20) to (80, 70)
Predicted box:
(30, 25) to (90, 60)


1. Area of Ground Truth
width = 80 - 20 = 60  
height = 70 - 20 = 50  
Area = 60 × 50 = 3000


2. Area of Predicted Box
width = 90 - 30 = 60  
height = 60 - 25 = 35  
Area = 60 × 35 = 2100

3. Intersection Coordinates
x_min = max(20, 30) = 30  
y_min = max(20, 25) = 25  
x_max = min(80, 90) = 80  
y_max = min(70, 60) = 60  

4. Intersection Area
width = 80 - 30 = 50  
height = 60 - 25 = 35  
Area = 50 × 35 = 1750

5. Union Area
Union = 3000 + 2100 - 1750 = 3350

6. IoU
IoU = 1750 / 3350 ≈ 0.52

Exercise 3- Challenge Problem

1. Bounding Box Issues

Cracks are long and thin, but bounding boxes are rectangular
This means most pixels inside the box will be road, not crack
Only a small percentage (maybe 5–20%) of pixels represent the actual crack, making it inefficient


2. Alternative Approach

A better approach is Semantic Segmentation  
It labels each pixel, so it can accurately follow the shape of the crack


3. R-CNN Suitability

Faster R-CNN is not ideal for this task
It works well for objects with clear shapes, but cracks are irregular and thin
Segmentation models are better suited for this problem
