Objective:

Exercise 1- PCA for Visualization

The Iris dataset has 4 features, which cannot be visualized directly
We will use PCA to reduce the data to 2 dimensions and visualize it

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import pandas as pd
```

```python
iris = load_iris()
```

```python
X = pd.DataFrame(iris.data, columns=iris.feature_names)
y = pd.Series(iris.target)
```

```python
X.head()
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
pca = PCA(n_components=2)
principal_components = pca.fit_transform(X_scaled)
```

```python
pca_df = pd.DataFrame(
    principal_components,
    columns=['PC1', 'PC2']
)
```

```python
pca_df['Target'] = y
```

```python
plt.figure(figsize=(8, 6))
scatter = plt.scatter(
    pca_df['PC1'],
    pca_df['PC2'],
    c=pca_df['Target']
)

plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA Visualization of Iris Dataset')
plt.show()
```

The 2D PCA projection separates the three iris classes reasonably well.
One class is clearly separated, while the other two show some overlap.
Overall, PCA provides a good low-dimensional representation of the data.
