Exercise 1- Conceptual Questions

1. The Markov Property

In Blackjack, the state defined by the player's current hand total, the dealer's visible card, and whether the player has a usable ace satisfies the Markov Property. This is because all necessary information to make an optimal decision is contained in the current state, and past states do not provide additional useful information

In Poker, defining a Markov state is much more difficult because of hidden information. The current visible state does not fully represent the environment, so past actions and observations may still be relevant


2. The Discount Factor (γ)

Agent A (γ = 0.1) heavily prioritizes immediate rewards and will likely choose a shorter but possibly suboptimal path, focusing less on long-term outcomes

Agent B (γ = 0.99) values future rewards much more and will try to maximize total long-term reward, making it more likely to find the optimal (shortest) path to the exit

Therefore, Agent B is more likely to find the absolute shortest path


3. Policies

A stochastic policy is better in situations where unpredictability is important, such as in games like Rock-Paper-Scissors. If the agent always chooses the same action (deterministic), it becomes predictable and exploitable.

A stochastic policy prevents opponents from exploiting patterns by introducing randomness into decisions

Exercise 2- The Bellman Equation


- γ = 0.9
- Policy: π(left) = 0.5, π(right) = 0.5


1. Value of Terminal States

Vπ(s1) = 0  
Vπ(s3) = 0  
Explanation: Terminal states have no future rewards



2. Value of Start State Vπ(s2)

Bellman Equation:
Vπ(s2) = Σ π(a|s2) * [R(s2, a) + γ * V(s')]

For action LEFT:
- Reward = +10
- Next state = s1 → V(s1) = 0

For action RIGHT:
- Reward = -10
- Next state = s3 → V(s3) = 0


Calculation
Vπ(s2) = 0.5 * (10 + 0.9 * 0) + 0.5 * (-10 + 0.9 * 0)
Vπ(s2) = 0.5 * 10 + 0.5 * (-10)
Vπ(s2) = 5 - 5 = 0

Exercise 3- Policy Evaluation

π'(left) = 0.8  
π'(right) = 0.2  


1. Calculate Vπ'(s2)

Vπ'(s2) = 0.8 * (10 + 0.9 * 0) + 0.2 * (-10 + 0.9 * 0)
Vπ'(s2) = 0.8 * 10 + 0.2 * (-10)
Vπ'(s2) = 8 - 2 = 6


2. Comparison

Previous value: Vπ(s2) = 0  
New value: Vπ'(s2) = 6  

The new policy is better because it chooses the action leading to the positive reward more often

This makes intuitive sense because going left leads to +10, while going right leads to -10.Increasing the probability of going left increases the expected reward
