RAG vs Fine-tuning for Customer Support

In this case, RAG is clearly the better choice.
RAG (Retrieval-Augmented Generation)
Pros:
-Always uses up-to-date documentation
-No need to retrain model weekly
-Easy to scale and update knowledge base
-Reduces hallucinations by grounding answers in real docs

Cons-
-Requires vector database infrastructure
-Slightly higher latency due to retrieval step


Fine-tuning
Pros:
-More natural responses in company style
-Faster inference (no retrieval step)

Cons:
-Expensive to retrain weekly
-Quickly becomes outdated
-Risk of memorization instead of true retrieval


Conclusion-
Since documentation changes weekly, RAG is much more efficient and maintainable than constant fine-tuning.


3. Importance of the "Thought" Step in ReAct
The "Thought" step is important because it allows the model to explicitly reason before taking an action.

Benefits-
-Makes reasoning process transparent
-Helps debug failed agent behavior
-Allows developers to see WHY an action was chosen
-Reduces random or incorrect tool usage
-Improves multi-step reasoning performance

Debugging advantage-
If an agent fails, we can inspect the "Thought" trace to understand whether:
-The reasoning was incorrect
-The wrong tool was selected
-The model misunderstood the question

Without this step, failures would look like black-box outputs with no explanation.

Exercise 2 — ReAct Prompt Design


You MUST follow the ReAct format:

Thought: think about what to do next
Action: choose a tool and input
Observation: result from tool

You have access to the following tools:

1. search(query: string)
-Use this to search general information on the web.
-Format:
Action: search(query="...")

2. get_movie_details(title: string)
-Use this to get details about a movie including director.
-Format:
Action: get_movie_details(title="...")



Example (Few-shot)
Question: Who was the US president when the first iPhone was released?

Thought: I need to find when the first iPhone was released.
Action: search(query="first iPhone release date")
Observation: The first iPhone was released in 2007.

Thought: Now I need to find who was US president in 2007.
Action: search(query="US president 2007")
Observation: George W. Bush was US president in 2007.

Thought: I now have the final answer.
Final Answer: George W. Bush

Exercise 3— Designing a RAG System
Ingestion Process (Vector Database Setup)

1. Collect all documents:
   -50 PDF lecture notes
   -10 textbook chapters
   -personal notes (text files)
2. Extract text:
   -Use PDF parser (e.g., PyMuPDF or pdfminer)
   -Clean formatting issues
3. Chunking:
   -Split text into small overlapping chunks (e.g., 300–800 tokens)
   -Overlap ensures context is preserved
4. Embedding:
   -Convert each chunk into vector embeddings using a model like OpenAI embeddings or SentenceTransformers
5. Store in vector database:
   -Save embeddings + metadata (lecture number, chapter, topic) in FAISS / Pinecone / Weaviate

Retrieval Process (Query Understanding)
   User query:
"Can you explain the Bellman equation in the context of the cliff walking example from Lecture 4?"

Steps:
1. Convert query into embedding vector
2. Compare with stored document embeddings using similarity search
3. Retrieve top-k relevant chunks:
   -Lecture 4 notes on reinforcement learning
   -Sections mentioning Bellman equation
   -Cliff walking example description
4. Optionally re-rank results for best relevance
5. Pass selected chunks to LLM for final answer generation

3. Generation Process (Final Prompt to LLM)
You are a study assistant for a university course.
Use the provided context to answer the question accurately.
If the answer is not in the context, say I don't have enough information in the provided materials.
