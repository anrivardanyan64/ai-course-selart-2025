1. Why is text preprocessing more complex than numerical preprocessing

Text data is unstructured and ambiguous.
Words can have multiple meanings, different grammatical forms, and include punctuation, slang, or spelling variations
Numerical data usually only requires simple transformations like scaling or normalization.
In contrast, text must be tokenized, cleaned, normalized, and converted into a numerical representation before machine learning models can use it

2. Stop Word Dilemma

Removing stop words can sometimes remove important meaning

For example, phrases like "not true" or "not correct" contain the word "not", which is often considered a stop word. 
If "not" is removed, the sentence meaning changes completely

"This is not true" → "true"

This could cause a fake news detection model to misunderstand the meaning of the sentence and produce incorrect predictions

3. Stemming vs Lemmatization

I would argue for stemming

Stemming is computationally faster than lemmatization because it simply removes word endings without performing full linguistic analysis
In a real-time search engine processing millions of queries per second, speed is extremely important
The trade-off is that stemming may produce less accurate or less readable words, but the gain in performance makes it suitable for large-scale systems

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
```

```python
import nltk

nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
stop_words = set(stopwords.words("english"))
```

```python
def preprocess_text(raw_text):

    tokens = word_tokenize(raw_text)
    tokens = [token.lower() for token in tokens]
    tokens = [token for token in tokens if token.isalpha()]
    tokens = [token for token in tokens if token not in stop_words]

    return tokens
```

```python
import spacy
from nltk.stem import PorterStemmer
```

```python
nlp = spacy.load("en_core_web_sm")
stemmer = PorterStemmer()
```

```python
text = """
Artificial intelligence (AI) is intelligence demonstrated by machines, 
in contrast to the natural intelligence displayed by humans and animals. 
AI research has been defined as the study of intelligent agents.
These agents perceive their environment and take actions that maximize 
their chances of achieving goals.
"""

doc = nlp(text)

tokens = [
    token.text.lower()
    for token in doc
    if token.is_alpha and not token.is_stop
]

lemmatized = [
    token.lemma_.lower()
    for token in doc
    if token.is_alpha and not token.is_stop
]
```

```python
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()

stemmed = [stemmer.stem(token) for token in tokens]
```

```python
print("Lowercase tokens:", len(tokens))
print("Unique:", len(set(tokens)))

print()

print("Stemmed tokens:", len(stemmed))
print("Unique:", len(set(stemmed)))

print()

print("Lemmatized tokens:", len(lemmatized))
print("Unique:", len(set(lemmatized)))
```

Stemming usually produces the smallest vocabulary because it aggressively removes word endings and merges many word forms into the same root
However, the resulting words are often not real English words and can look unnatural

Lemmatization usually produces more readable and human-like tokens because it converts words into their proper dictionary form

Therefore:
Stemming reduces vocabulary size the most
Lemmatization produces the most readable tokens
