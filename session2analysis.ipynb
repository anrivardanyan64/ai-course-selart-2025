```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("car_data.csv")
```

```python
print("First 5 rows:")
print(df.head(), "\n")

print("Info:")
print(df.info(), "\n")

print("Statistical Summary:")
print(df.describe(), "\n")
```

```python
average_price = df["price_usd"].mean()
print(f"The average car price is: ${average_price:,.2f}")
```

```python
powerful_cars = df[df["hp "] > 300]
print("\nCars with more than 300 horsepower:")
print(powerful_cars)
```

```python
plt.figure(figsize=(8, 5))
plt.bar(df["model"], df["price_usd"], color='skyblue', edgecolor='black')
plt.title("Price of Car Models")
plt.xlabel("Car Model")
plt.ylabel("Price (USD)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(df["hp "], df["price_usd"], color='green')
plt.title("Horsepower vs. Price")
plt.xlabel("Horsepower (hp)")
plt.ylabel("Price (USD)")
plt.grid(True)
plt.tight_layout()
plt.show()
```
