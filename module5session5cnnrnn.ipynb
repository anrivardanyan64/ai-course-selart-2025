2. Convolution with Vertical Edge Filter

Kernel
1  0 -1  
1  0 -1  
1  0 -1

Top-left position (centered at row 2, column 2)

Image patch covered by filter
0 0 0  
0 0 0  
0 0 0  

Element-wise multiplication:
(0*1) + (0*0) + (0*-1) + (0*1) + (0*0) + (0*-1) + (0*1) + (0*0) + (0*-1)

Sum = 0
Output value = 0

Top-right position (centered at row 2, column 4)

Image patch
0 10 10  
0 10 10  
0 10 10  

Now multiply
Row 1: (0*1) + (10*0) + (10*-1) = -10  
Row 2: (0*1) + (10*0) + (10*-1) = -10  
Row 3: (0*1) + (10*0) + (10*-1) = -10  

Sum = -30
Output value = -30

3. Interpretation

A high positive or negative value indicates a strong vertical edge.

A value close to zero means.
There is no vertical edge detected in that region.

In our case:
- Left region → 0 (no edge)
- Right region → strong edge detected

Exercise 2- RNN Hidden State

Given
W_hh = 0.5  
W_xh = 1.5  
b = 0  

Formula
h_t = tanh(W_hh * h_{t-1} + W_xh * x_t)

t = 0

h_{-1} = 0  
x_0 = 1  
h_0 = tanh(0.5*0 + 1.5*1)
h_0 = tanh(1.5)
h_0 ≈ 0.905

t = 1

x_1 = 2  
h_1 = tanh(0.5*h_0 + 1.5*2)
h_1 = tanh(0.5*0.905 + 3)
h_1 = tanh(0.4525 + 3)
h_1 = tanh(3.4525)
h_1 ≈ 0.998

t = 2

x_2 = -3  
h_2 = tanh(0.5*h_1 + 1.5*(-3))
h_2 = tanh(0.5*0.998 - 4.5)
h_2 = tanh(0.499 - 4.5)
h_2 = tanh(-4.001)
h_2 ≈ -0.999

Does h_2 depend on x_0?
Yes.

Because:
h_2 depends on h_1
h_1 depends on h_0
h_0 depends on x_0

Therefore, information flows through time via the hidden state.

Exercise 3- Architecture Strategy

1. Project A (Audio → Text)

Use: RNN (or LSTM)
Reason:
Audio is sequential data, and RNNs are designed to handle time sequences.


2. Project B (Satellite Images)

Use: CNN
Reason:
Images contain spatial patterns.
CNNs are designed to detect spatial features like edges and textures.


3. Project C (Customer Tabular Data)

Use: MLP
Reason:
This is structured tabular data without spatial or temporal structure.
MLPs work well for this type of problem.
