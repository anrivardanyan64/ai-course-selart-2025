TF-IDF Intuition

The word "explosion" will have a higher IDF score.

"Tuesday" appears very frequently in many news articles, so it is not informative.
"Explosion" is rarer and more specific, so it gets a higher IDF value.

2. Cosine Similarity vs Euclidean Distance

Cosine similarity is preferred because it focuses on the direction of the vectors, not their magnitude.
Longer documents naturally have larger values, which can distort Euclidean distance.
Cosine similarity normalizes this and compares the content rather than the size of the document.

3. Embeddings vs TF-IDF
Word embeddings are better.

TF-IDF treats words independently and cannot capture semantic similarity.

Word embeddings place similar words like "feline" and "kitten" close together in vector space, 
so they can recognize meaning.

```python
from sklearn.feature_extraction.text import TfidfVectorizer

corpus = [
    'the quick brown fox jumps over the lazy dog',
    'a brown dog was chasing a brown cat',
    'the lazy cat sleeps all day'
]

vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix = vectorizer.fit_transform(corpus)
print(vectorizer.get_feature_names_out())
print(tfidf_matrix.toarray())
```

The word "brown" appears in the first and second sentences, 
so it has non-zero values in those rows.

It does not appear in the third sentence, 
so its value is zero in the third row.

This shows how TF-IDF represents the importance of words in each document.

```python
import gensim.downloader as api

word_vectors = api.load("glove-wiki-gigaword-50")
print(word_vectors.most_similar("woman", topn=5))
print()
print(word_vectors.most_similar("car", topn=5))
```

```python
print(word_vectors.most_similar(
    positive=['king', 'woman'],
    negative=['man']
))
```

```python
print(word_vectors.most_similar(
    positive=['france', 'berlin'],
    negative=['paris']
))
```

```python
print("cat vs dog:", word_vectors.similarity("cat", "dog"))
print("cat vs car:", word_vectors.similarity("cat", "car"))
```

The model successfully finds similar words and solves analogies.

For example:
"cat" and "dog" have higher similarity than "cat" and "car"
The analogy "king - man + woman" gives "queen"

This shows that the model captures semantic meaning.
Word embeddings understand relationships between words, 
not just their frequency, making them powerful for NLP tasks.
