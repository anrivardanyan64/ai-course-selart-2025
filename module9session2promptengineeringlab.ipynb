Exercise 1— Conceptual Questions

1. Scaling Laws
Doubling dataset size usually gives better returns than doubling model size, especially when the model is not yet data-saturated. More data improves generalization and reduces overfitting, while just increasing parameters without enough data leads to inefficient learning
2. Pre-training Objective
Even though next-word prediction is simple, to perform it well the model must learn grammar, facts, reasoning patterns, relationships between concepts, and real-world knowledge embedded in text. These implicit structures allow emergent abilities like reasoning and coding
3. In-Context Learning vs Fine-tuning
With small data, in-context learning is safer because fine-tuning may overfit and degrade general performance. In-context learning allows flexibility and avoids retraining costs, but fine-tuning gives more consistent style replication if enough data is available

Exercise 3— Chain of Thought Prompting

Standard Prompt Attempt
A bat and a ball cost $1.10 in total. The bat costs $1.00 more than the ball. How much does the ball cost?

Chain-of-Thought Prompt
A bat and a ball cost $1.10 in total. The bat costs $1.00 more than the ball. How much does the ball cost? Let's think step by step

Analysis
Chain-of-Thought prompting is more effective because it forces the model to decompose the problem into intermediate steps. This reduces reliance on intuitive but incorrect shortcuts and improves logical consistency, leading to correct arithmetic reasoning instead of pattern-based guessing
